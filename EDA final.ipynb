```python
print("Missing values in each column:\n")
print(df.isnull().sum())
```

```python
df['High_Revenue'] = (df['Revenue'] > df['Revenue'].mean()).astype(int)
df['High_Sales'] = (df['Sales'] > df['Sales'].mean()).astype(int)
```

```python
df['Profit_Margin'] = df['Profit'] / (df['Revenue'] + 1e-5)
df['Revenue_per_Unit'] = df['Revenue'] / (df['Quantity'] + 1e-5)
```

```python
df['Gender'] = df['Gender'].replace({'Genderfluid': 'Other', 'Genderqueer': 'Other', 'Agender': 'Other'})
```

```python
north_states = ['Delhi', 'Chandigarh']
south_states = ['Chennai', 'Hyderabad', 'Bangalore']
west_states = ['Mumbai', 'Pune']

def assign_region(state):
    if state in north_states:
        return 'North'
    elif state in south_states:
        return 'South'
    elif state in west_states:
        return 'West'
    else:
        return 'Other'

df['Region'] = df['State'].apply(assign_region)
```

```python
df['Date of Order'] = pd.to_datetime(df['Date of Order'], errors='coerce')

df['Order_Month'] = df['Date of Order'].dt.month
df['Order_Year'] = df['Date of Order'].dt.year
df['Order_Day'] = df['Date of Order'].dt.day
df['Order_Weekday'] = df['Date of Order'].dt.day_name()
```

```python
category_sales = df.groupby('Product Category')['Sales'].mean().reset_index()
category_sales.columns = ['Product Category', 'Avg_Category_Sales']

df = df.merge(category_sales, on='Product Category', how='left')
```

```python
df = pd.get_dummies(df, columns=['Mode of Payment'], prefix='Mode', drop_first=True)
```

```python
print(df['Return Reason'].value_counts())

def simplify_reason(reason):
    if 'Qual' in reason:
        return 'Quality'
    elif 'Defective' in reason:
        return 'Defect'
    elif 'High Cost' in reason:
        return 'Cost'
    else:
        return 'Other'

df['Return_Reason_Grouped'] = df['Return Reason'].apply(simplify_reason)
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
print(df.columns.tolist())
```
